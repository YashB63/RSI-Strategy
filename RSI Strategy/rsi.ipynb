```python
def rsi_calculate(asset):
    df = yf.download(asset,start='2013-1-1')# let's go 10 years back
    df['Moving_Averages_200'] = df['Adj Close'].rolling(window=200).mean()
    df['Price_Change'] = df['Adj Close'].pct_change()
    df['Upmove'] = df['Price_Change'].apply(lambda x: x if x>0 else 0)
    df['Downmove'] = df['Price_Change'].apply(lambda x: abs(x) if x<0 else 0)
    df['Average_Up'] = df['Upmove'].ewm(span=19).mean()
    df['Average_Down'] = df['Downmove'].ewm(span=19).mean()
    df = df.dropna()
    df['RS'] = df['Average_Up']/df['Average_Down']
    df['RSI'] = df['RS'].apply(lambda x: 100-(100/(x+1)))
    df.loc[(df['Adj Close'] > df['Moving_Averages_200']) & (df['RSI']<30), 'Buy'] = 'Yes'
    df.loc[(df['Adj Close'] > df['Moving_Averages_200']) | (df['RSI']>30), 'Buy'] = 'No'
    return df
```

```python
def getSignals(df):
    Buying_Dates = []
    Selling_Dates = []
    
    for i in range(len(df)):
        if "Yes" in df['Buy'].iloc[i]:#iloc to check for the row value after iterating every single row
            #append row name to buying dates
            Buying_Dates.append(df.iloc[i+1].name)#+1 as we want to buy on the next day
            #Now more tricky one is to define the selling date as we have to account for RSI not exceeding 40 over th next 10 trading days
            for j in range(1,11):
                if df['RSI'].iloc[i+j]>40:
                    Selling_Dates.append(df.iloc[i+j+1].name)
                    break
                elif j==10:
                    Selling_Dates.append(df.iloc[i+j+1].name)
    
    return Buying_Dates,Selling_Dates 
            
```

```python
frame = rsi_calculate(tickers[0])
buy, sell = getSignals(frame)
```
